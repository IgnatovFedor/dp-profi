```python
from pathlib import Path
from random import choice

import pandas as pd

data_path = Path('data')

df_1 = pd.read_csv(data_path / '1.csv')
df_2 = pd.read_csv(data_path / '2.csv')
# df_3 = pd.read_csv(data_path / '3.csv')  # будет доступно только после 10:00 15.04.21

topics = df_1['topic'].unique()


def predict(message: str) -> str:
    """Предсказываем топик основываясь на тексте сообщения"""
    return choice(topics)


df_2['topic'] = df_2.apply(lambda row: predict(row['message']), axis=1)

df_2.to_csv(data_path / 'out.csv', index=False)

df_2
```
